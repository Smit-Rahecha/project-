```python

t = (10, 20, 30, 20, 40)

# 1. count()
print("1. count():", t.count(20))

# 2. index()
print("2. index():", t.index(30))

# 3. len()
print("3. len():", len(t))

# 4. max()
print("4. max():", max(t))

# 5. min()
print("5. min():", min(t))
```

```python

s = {10, 20, 30}

print("Original Set:", s)

# 1. add()
s.add(40)
print("1. add():", s)

# 2. remove()
s.remove(20)
print("2. remove():", s)

# 3. discard()
s.discard(100)
print("3. discard():", s)

# 4. pop()
s.pop()
print("4. pop():", s)

# 5. clear()
s.clear()
print("5. clear():", s)
```
