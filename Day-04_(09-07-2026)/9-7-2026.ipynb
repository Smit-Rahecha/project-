```python
import pandas as pd
import numpy as np

# Load the sustainability dataset
df = pd.read_csv('/content/renewable_energy_sustainability_dataset.csv')

# Display first 5 rows
df.head()
```

```python
# Check the number of rows and columns
print("Dataset Shape:", df.shape)

# Display column names
print("\nColumn Names:")
print(df.columns.tolist())

# Check data types and non-null values
print("\nDataset Information:")
df.info()

# Statistical summary of numerical columns
print("\nStatistical Summary:")
print(df.describe())
```

```python
# Check missing/null values in each column
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Percentage of missing values in each column
missing_percentage = (df.isnull().sum() / len(df)) * 100

print("Missing Value Percentage:")
print(missing_percentage)
```

```python
# Handle missing values

# Fill numerical columns with median
numerical_columns = df.select_dtypes(include=['number']).columns

for column in numerical_columns:
    df[column] = df[column].fillna(df[column].median())


# Fill categorical columns with mode
categorical_columns = df.select_dtypes(include=['object']).columns

for column in categorical_columns:
    if not df[column].mode().empty:
        df[column] = df[column].fillna(df[column].mode()[0])


# Verify that missing values have been handled
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
# Analyze each column individually

for column in df.columns:
    print("\n" + "=" * 50)
    print("COLUMN:", column)
    print("Data Type:", df[column].dtype)
    print("Unique Values:", df[column].nunique())

    if pd.api.types.is_numeric_dtype(df[column]):
        print("Mean:", df[column].mean())
        print("Median:", df[column].median())
        print("Minimum:", df[column].min())
        print("Maximum:", df[column].max())
    else:
        print("Most Frequent Value:")
        print(df[column].value_counts().head())
```

```python
# Select numerical columns
numeric_df = df.select_dtypes(include=['number'])

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()

print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Find relationships between numerical features

columns = numeric_df.columns

for i in range(len(columns)):
    for j in range(i + 1, len(columns)):

        col1 = columns[i]
        col2 = columns[j]

        correlation = numeric_df[col1].corr(numeric_df[col2])

        print(f"{col1} vs {col2}: Correlation = {correlation:.2f}")
```
