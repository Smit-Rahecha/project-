```python
import pandas as pd
import numpy as np

def generate_student_data(num_samples=1000):
    # Define possible values for categorical columns
    stream = ['CS', 'Mech', 'EC', 'IT', 'Civil']
    genders = ['F', 'M']

    np.random.seed(42)
    # Generate random choices for Degree and Gender
    data = pd.DataFrame({
        'Stream': np.random.choice(stream, num_samples),
        'Gender': np.random.choice(genders, num_samples)
    })

    data['Maths']  = np.round(np.clip(np.random.normal(6, 1, 1000), 1, 10))
    data['English']  = np.round(np.clip(np.random.normal(7, 1, 1000), 1, 10))
    data['Science']  = np.round(np.clip(np.random.normal(5, 1, 1000), 1, 10))
    # Calculate the 'Total' score
    data['Total'] = data['Maths'] + data['English'] + data['Science']

    # Define a function to determine the 'Grade' based on the 'Total' score
    def assign_grade(total):
        if total >= 24:
            return 'A'
        elif total >= 18:
            return 'B'
        elif total >= 15:
            return 'C'
        else:
            return 'D'

    # Apply the grading logic to create the 'Grade' column
    data['Grade'] = data['Total'].apply(assign_grade)
    data['Fuel'] = np.round(np.clip(np.random.normal(10.5, 3.16, 1000), 1, 20),2)
    #data['Lonavala'] = np.random.choice([0,1], size = 1000, p=[0.3, 0.7])

    return data

# Generate 1000 samples
df = generate_student_data(num_samples=1000)

# Display the first 5 rows of the generated DataFrame
print(df.head())

# Optional: Save the data to a CSV file
# df_synthetic.to_csv('synthetic_student_data.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df = pd.read_csv(filename)

df.head()
```

```python
# The dataset is loaded using Pandas.
# df.head() displays the first 5 rows to understand its structure,
# features, values, and overall format.
```

```python
print("Dataset Shape:", df.shape)
print("\nColumn Names:")
print(df.columns.tolist())

print("\nDataset Information:")
df.info()

print("\nStatistical Summary:")
display(df.describe(include='all'))
```

```python
print("Duplicate Rows:", df.duplicated().sum())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
numeric_df = df.select_dtypes(include=np.number)

correlation_matrix = numeric_df.corr()

display(correlation_matrix)
```

```python
plt.figure(figsize=(12, 8))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x=numerical_cols[0], kde=True)
plt.title(f'Distribution of {numerical_cols[0]}')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x=numerical_cols[0])
plt.title(f'Box Plot of {numerical_cols[0]}')
plt.show()
```

```python
if len(numerical_cols) >= 2:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(
        data=df,
        x=numerical_cols[0],
        y=numerical_cols[1]
    )
    plt.title(
        f'Relationship Between {numerical_cols[0]} and {numerical_cols[1]}'
    )
    plt.show()
```

```python
if len(categorical_cols) > 0:
    plt.figure(figsize=(12, 6))
    sns.countplot(
        data=df,
        x=categorical_cols[0],
        order=df[categorical_cols[0]].value_counts().head(10).index
    )
    plt.xticks(rotation=45)
    plt.title(f'Top Categories in {categorical_cols[0]}')
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_scaled = df.copy()

df_scaled[numerical_cols] = scaler.fit_transform(
    df[numerical_cols]
)

display(df_scaled.head())
```

```python
df_encoded = pd.get_dummies(
    df_scaled,
    columns=categorical_cols,
    drop_first=True
)

display(df_encoded.head())

print("Shape after encoding:", df_encoded.shape)
```

```python
label_encoder = LabelEncoder()

for col in categorical_cols:
    df_scaled[col] = label_encoder.fit_transform(
        df_scaled[col].astype(str)
    )
```
